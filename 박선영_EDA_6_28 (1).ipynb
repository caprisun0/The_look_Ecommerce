```python
# 연도별 평균 WAU 증가량 구하기
wau_yearly = wau.resample('Y').mean()

# 3. Calculate the annual average WAU increase
wau_yearly_increase = wau_yearly.diff().dropna()

# Optional: Convert the yearly average WAU to a DataFrame for better readability
wau_yearly_df = wau_yearly.reset_index()
wau_yearly_df.columns = ['year', 'average_wau']

wau_yearly_increase_df = wau_yearly_increase.reset_index()
wau_yearly_increase_df.columns = ['year', 'annual_increase']

# Display the results
print("연도별 평균 WAU:")
print(wau_yearly_df)
print("\n연도별 평균 WAU 증가:")
print(wau_yearly_increase_df)
```

```python
order_items['created_at'] = pd.to_datetime(order_items['created_at'], errors='coerce')
# Calculate WAU (Weekly Active Users)
weekly_order = order_items.set_index('created_at').groupby(pd.Grouper(freq='W-MON'))['order_id'].nunique()

weekly_order_yearly = weekly_order.resample('Y').mean()

weekly_order_increase = weekly_order_yearly.diff().dropna()
weekly_order_yearly_df = weekly_order_yearly.reset_index()
weekly_order_yearly_df.columns = ['year', 'average_order']

weekly_order_yearly_increase_df = weekly_order_increase.reset_index()
weekly_order_yearly_increase_df.columns = ['year', 'annual_increase']

# Display the results
print("연도별 평균 주문수:")
print(weekly_order_yearly_df)
print("\n연도별 평균 주문수 증가:")
print(weekly_order_yearly_increase_df)
```

```python
first_order_occurrence = order_items.groupby('order_id')['created_at'].min().reset_index()

# 2. Extract the year from 'created_at'
first_order_occurrence['year'] = first_order_occurrence['created_at'].dt.year

# 3. Count the number of unique orders per year
orders_per_year = first_order_occurrence.groupby('year')['order_id'].nunique().reset_index()
orders_per_year.columns = ['year', 'orders']

# Display the results
print("연도별 주문수:")
print(orders_per_year)

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(orders_per_year['year'], orders_per_year['orders'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.title('연도별 주문수')
plt.xticks(orders_per_year['year'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
order_items['year'] = order_items['created_at'].dt.year

# 2. Calculate the total sales per year
sales_per_year = order_items.groupby('year')['sale_price'].sum().reset_index()
sales_per_year.columns = ['year', 'total_sales']

# Display the results
print("연도별 매출:")
print(sales_per_year)

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(sales_per_year['year'], sales_per_year['total_sales'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('총 매출')
plt.title('연도별 매출')
plt.xticks(sales_per_year['year'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
weekly_sale = order_items.set_index('created_at').groupby(pd.Grouper(freq='W-MON'))['sale_price'].sum()

# 2. Calculate the average weekly sales per year
weekly_sale_yearly = weekly_sale.resample('Y').mean()

# 3. Calculate the annual increase in average weekly sales
weekly_sale_increase = weekly_sale_yearly.diff().dropna()

# Convert the results to DataFrames for better readability
weekly_sale_yearly_df = weekly_sale_yearly.reset_index()
weekly_sale_yearly_df.columns = ['year', 'average_weekly_sales']

weekly_sale_yearly_increase_df = weekly_sale_increase.reset_index()
weekly_sale_yearly_increase_df.columns = ['year', 'annual_increase']

# Display the results
print("연도별 평균 매출:")
print(weekly_sale_yearly_df)
print("\n연도별 평균 매출 증가:")
print(weekly_sale_yearly_increase_df)
```
